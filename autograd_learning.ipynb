```python
import torch
from torchvision.models import resnet18, ResNet18_Weights

# resnet18 is a pretrained model
model = resnet18(weights=ResNet18_Weights.DEFAULT)

# create 1 random tensor with channel 3 (r, g, b) and aspect ratio 64 x 64
data = torch.rand(1, 3, 64, 64)

# give the data a random label
labels = torch.rand(1, 1000)
```

```python
#print(model)
```

```python
# perform forward pass on this data
prediction = model(data)
```

```python
# calculate the loss
loss = (prediction - labels).sum()

# backward pass
loss.backward() 
# backward pass propogates the error through the network.
# stores the gradients for each parameter in the parameters .grad attribute
```

```python
# optimizer
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)

# initiate gradient descent
optim.step()
# adjusts each parameter by it's gradient times the learning rate
```
